# Analyzing data

## Number of releases

Finding the number of books released after January 1, 2000.

```python
# Querying database
query = \
    '''
    SELECT
        COUNT(book_id) AS number_of_books
    FROM
        books
    WHERE
        publication_date :: date >= '2000-01-01';
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

821 books have been released after January 1, 2000.

## Number of reviews and average rating

Finding the number of user reviews and the average rating for each book.

```python
# Querying database
query = \
    '''
    SELECT
        books.book_id, books.title, COUNT(reviews.review_id) AS number_of_reviews, AVG(ratings.rating) AS average_rating
    FROM
        books 
        INNER JOIN reviews ON books.book_id = reviews.book_id
        INNER JOIN ratings ON books.book_id = ratings.book_id
    GROUP BY
        books.book_id
    ORDER BY
        number_of_reviews, average_rating;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

```python
# Plotting the distribution of the number of reviews
plt.figure(figsize=(12,5))
plt.hist(data_query['number_of_reviews'], bins=200)
plt.title('Number of reviews histogram')
plt.ylabel('Number of books')
plt.xlabel('Number of reviews')
plt.show()

plt.figure(figsize=(12,5))
plt.hist(data_query['number_of_reviews'], bins=500)
plt.title('Number of reviews histogram - zoomed in')
plt.ylabel('Number of books')
plt.xlabel('Number of reviews')
plt.xlim(0,50)
plt.show()
```

The distribution of the number of reviews per books is sharply peaked at 1 review and is positively skewed, where only a small number of books have more than 50 reviews, but some of these outliers reach up to hunderds of reviews.

```python
# Plotting the distribution of the average_rating
plt.figure(figsize=(12,5))
plt.hist(data_query['average_rating'], bins=200)
plt.title('Average rating histogram')
plt.ylabel('Number of books')
plt.xlabel('Average rating ')
plt.show()
```

The distribution of the average book ratings is approximately normal, with a peak at a rating of 4. The distribution is shifted to the right, such that more books get the maximal ratings than the minimal ratings. We can also see that people tend to give books round ratings, in particular multiples of 0.5.

## Top publisher by number of books

Identifying the publisher that has released the greatest number of books with more than 50 pages (this will help exclude brochures and similar publications from the analysis).

```python
# Querying database
query = \
    '''
    SELECT
        books.publisher_id, publishers.publisher, COUNT(books.book_id) AS number_of_books
    FROM
        books
        INNER JOIN publishers ON books.publisher_id = publishers.publisher_id
    WHERE
        books.num_pages > 50
    GROUP BY
        books.publisher_id, publishers.publisher
    ORDER BY
        number_of_books DESC
    LIMIT
        1;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

The publisher that has released the greatest number of books with more than 50 pages is Penguin Books.

## Top author by average rating

Identifying the author with the highest average book rating (looking only at books with at least 50 ratings).

```python
# Querying database
# This query will be used as a subquery later, and returns the books with at least 50 ratings
query = \
    '''
    SELECT
        books.book_id
    FROM
        books
        INNER JOIN ratings ON books.book_id = ratings.book_id
    GROUP BY
        books.book_id
    HAVING
        COUNT(ratings.rating_id) >= 50;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

```python
# Querying database
# This query presents more information about the books with at least 50 ratings
query = \
    '''
    SELECT
        books.book_id, books.title, authors.author_id, authors.author, COUNT(ratings.rating_id) AS ratings_per_book
    FROM
        books
        INNER JOIN authors ON books.author_id = authors.author_id
        INNER JOIN ratings ON books.book_id = ratings.book_id
    GROUP BY
        books.book_id, books.title, authors.author_id, authors.author
    HAVING
        COUNT(ratings.rating_id) >= 50
    ORDER BY
        ratings_per_book DESC;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

```python
# Querying database
# Finding the author with the highest average book rating, while only considering books with at least 50 ratings
query = \
    '''
    SELECT
        authors.author_id, authors.author, AVG(ratings.rating) AS average_rating
    FROM
        books
        INNER JOIN authors ON books.author_id = authors.author_id
        INNER JOIN ratings ON books.book_id = ratings.book_id

    WHERE
        books.book_id IN 
            (
            SELECT
                books.book_id
            FROM
                books
                INNER JOIN ratings ON books.book_id = ratings.book_id
            GROUP BY
                books.book_id
            HAVING
                COUNT(ratings.rating_id) >= 50
            )

    GROUP BY
        authors.author_id
    ORDER BY
        average_rating DESC
    LIMIT
        1;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

The author with the highest average book rating (looking only at books with at least 50 ratings) is J.K. Rowling / Mary GrandPr.

## Average number of reviews

Finding the average number of text reviews among users who rated more than 50 books.

```python
# Querying database
# This query will be used as a subquery later, and returns the users who rated more than 50 books
query = \
    '''
    SELECT
        username
    FROM
        ratings
    GROUP BY
        username
    HAVING
        COUNT(rating_id) > 50;
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

```python
# Querying database
# Finding the average number of text reviews among users who rated more than 50 books
query = \
    '''
    SELECT
        COUNT(review_id) / COUNT(DISTINCT username) as average_reviews
    FROM
        reviews
    WHERE
        reviews.username in
        (
        SELECT
            username
        FROM
            ratings
        GROUP BY
            username
        HAVING
            COUNT(rating_id) > 50
        );
    '''

data_query = pd.io.sql.read_sql(query, con = engine)

# Presenting query results
display(data_query)
```

The average number of text reviews among users who rated more than 50 books is 24.
